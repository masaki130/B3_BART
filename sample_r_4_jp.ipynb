```python
# 再学習用のペアデータ作成
# 基本、sample_r_3_jp.ipynbのコピー
from espnet_phoneme_tokenizer import pyopenjtalk_g2p_prosody
# テキストファイルから日本語の文を読み込む
with open("output_1_4_0.txt", "r", encoding="utf-8") as file:     # wiki16,000文, output_1_1.txt
    japanese_texts = [line.strip() for line in file if line.strip()]
    print(japanese_texts)
```

```python
def read_phoneme_text_from_file(file_path, japanese_texts):
    with open(file_path, 'r', encoding='utf-8') as file:
        # 各行ごとに{"text": "テキスト", "phoneme_text": "フォネムテキスト"}という形式で囲んで、リストに格納
        phoneme_lines = [{'text': text, 'phoneme_text': line.strip()} for text, line in zip(japanese_texts, file.readlines())]
    return phoneme_lines

# インプットテキストファイルのパス
input_file_path = 'output_1_3_0.txt'  # 再学習用データ：output_1_3_0.txt, (評価用データ:output_1_3_0_0.txt)

# リスト内包表記を使って結果を生成
result = [{'text': text['text'], 'phoneme_text': text['phoneme_text']} for text in read_phoneme_text_from_file(input_file_path, japanese_texts)]

# 結果を表示
for item in result:
    print(item)
```

```python
# all_5.jsonとして保存, basic5000の日本語文と音素列のペア
import json
json.dump(result, open('all_5.json', 'w', encoding='utf-8'), ensure_ascii=False)
```
